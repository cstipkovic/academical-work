```python
im = io.imread('images-for-notebooks/empire.jpg')
print(im)
```

```python
im2 = 255 - im # L = 256
print(im2)
```

```python
im3 = (100.0/255) * im + 100 # limitando ao intervalo 100...200
print(im3)
```

```python
im4 = 255.0 * (im/255.0) ** 2 # quadrado
print(im4)
```

```python
plt.subplot(131), plt.imshow(im2, cmap='gray')
plt.subplot(132), plt.imshow(im3, cmap='gray')
plt.subplot(133), plt.imshow(im4, cmap='gray')
```

```python
x = np.arange(0, 256)
f1 = x
f2 = 255 - x
f3 = (100.0 / 255.0) * x + 100 # for i in x
f4 = 255.0 * (x / 255.0) ** 2
```

```python
plt.plot(x, f1, 'r')
plt.plot(x, f2, 'g')
plt.plot(x, f3, 'b')
plt.plot(x, f4, 'y')
plt.xlabel('Entrada')
plt.ylabel('Saida')
plt.legend(['f(x) = x', 'f(x) = 255 - x', 'f(x) = (100.0 / 255.0) * x + 100', 'f(x) = 255.0 * (x / 255.0) ** 2'])
```
